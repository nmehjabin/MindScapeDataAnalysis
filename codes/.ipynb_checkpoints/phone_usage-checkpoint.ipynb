```python
'''
This version looks into phone usage for each participant in cntx vs generic
'''
import pandas as pd

# --- Load and clean ---
df = pd.read_csv("feat_dataset_filtered_with_weeks.csv")
df.columns = df.columns.str.replace(' ', '_')

# --- Define your activity columns ---
activity_cols = [
    'unlock_duration_ep_0', 'unlock_num_ep_0'
]

# --- Keep only columns that exist ---
activity_cols = [c for c in activity_cols if c in df.columns]

# --- Compute per-participant averages for Week 1 and Week 6 ---
per_uid_week = (
    df[df['week_per_uid'].isin([1, 6,7,8])]
    .groupby(['uid', 'week_per_uid'])[activity_cols]
    .mean(numeric_only=True)
    .reset_index()
)

# --- Pivot to align Week 1 and Week 6 side-by-side ---
pivot = per_uid_week.pivot(index='uid', columns='week_per_uid', values=activity_cols)
pivot.columns = [f"{col}_w{wk}" for (col, wk) in pivot.columns]
pivot = pivot.reset_index()

# --- Save the raw week 1 & 6 averages ---
pivot.to_csv("phone.csv", index=False)
```

```python


import pandas as pd
import matplotlib.pyplot as plt

phone = pd.read_csv("phone.csv")
# print(phone.columns)

phone['unlock_duration_change_cntx'] = phone['unlock_duration_ep_0_w6'] - phone['unlock_duration_ep_0_w1']
phone['unlock_num_change_cntx'] = phone['unlock_num_ep_0_w6'] - phone['unlock_num_ep_0_w1']

phone['unlock_duration_change_generic'] = phone['unlock_duration_ep_0_w8'] - phone['unlock_duration_ep_0_w7']
phone['unlock_num_change_generic'] = phone['unlock_num_ep_0_w8'] - phone['unlock_num_ep_0_w7']

# summary1 = phone[[
#     'unlock_duration_ep_0_w1', 'unlock_duration_ep_0_w6',
#     'unlock_num_ep_0_w1', 'unlock_num_ep_0_w6'
# ]].agg(['mean', 'std']).T

# print("week 1 vs week 6 \n",summary1)

# summary2 = phone[[
#     'unlock_duration_ep_0_w1', 'unlock_duration_ep_0_w8',
#     'unlock_num_ep_0_w1', 'unlock_num_ep_0_w8'
# ]].agg(['mean', 'std']).T
# print("week 1 vs week 8 \n",summary2)


#graph for week cntx
plt.figure(figsize=(10,5))
plt.axhline(0, color='gray', linestyle='--')
plt.bar(phone['uid'], phone['unlock_duration_change_cntx'])
plt.xticks(rotation=90)
plt.ylabel('Change in Unlock Duration (cntx week)')
plt.title('Per-participant Change in Unlock Duration')
plt.tight_layout()
plt.show()


#graph for week generic
plt.figure(figsize=(10,5))
plt.axhline(0, color='gray', linestyle='--')
plt.bar(phone['uid'], phone['unlock_duration_change_generic'])
plt.xticks(rotation=90)
plt.ylabel('Change in Unlock Duration (generic week)')
plt.title('Per-participant Change in Unlock Duration')
plt.tight_layout()
plt.show()
```
