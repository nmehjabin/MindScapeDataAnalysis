```python
'''

Filtering feat_dataset based on demographics csv file to remove test user
Then labeling week in the dataset for each users from week 1 to week 8; based on each user start date of study

demographics_journaling : colum uid 
feat_dataset: filter the column uid based on the demographics_journaling.csv column uid 
'''

import pandas as pd

#load files
demo = pd.read_csv("~/Desktop/MaterialsbySubigya/Mindscape dataAnalysis_new/MindScape Data/demographics_journaling.csv")
feat = pd.read_csv("~/Desktop/MaterialsbySubigya/Mindscape dataAnalysis_new/MindScape Data/feat_dataset.csv")

filtered = feat[feat["uid"].isin(demo["uid"])]

#save the file 
filtered.to_csv("feat_dataset_filtered.csv")
# print(filtered.head())
print("pass")
```

```python
'''
now use the feat_dataset_filtered.csv file to label week and add the week label using column 'day'
note: there are values of 1969 as data so we are getting 'day' >= 2020
'''
import pandas as pd

# Load filtered dataset
df = pd.read_csv("feat_dataset_filtered.csv", low_memory=False)

# Convert day to  datetime
df["day"] = pd.to_datetime(df["day"])

df = df[df["day"].dt.year >= 2020]

# Sort by uid and day
df = df.sort_values(["uid", "day"])

# Compute each user's personal start date
df["start_day"] = df.groupby("uid")["day"].transform("min")

# Compute week number per uid
df["week_per_uid"] = ((df["day"] - df["start_day"]).dt.days // 7) + 1

# Save
df.to_csv("feat_dataset_filtered_with_weeks.csv", index=False)

print(df.head())
print(df[["uid", "day", "week_per_uid"]].head(20))
```
